Visualizing intermediate convnet outputs (intermediate activations)—
Useful for understanding how successive convnet layers transform their input, and for get-
ting a first idea of the meaning of individual convnet filters.

```python
from keras.models import load_model
model = load_model('cat_vs_dog_2.h5')   # importing the saved model
```

```python
model.summary()
```

```python
#you’ll get an input image—a picture of a cat, not part of the images the network was trained on
img_path ='/home/vajir/Desktop/Untitled Folder/ML/dog_vs_cat/dog_and_cat_small/test/cats/cat.1500.jpg'
from keras.preprocessing import image
import numpy as np
img = image.load_img(img_path ,target_size =(150 ,150))
img_tensor = image.img_to_array(img)            # converting into numpy array
img_tensor = np.expand_dims(img_tensor ,axis =0)  # expanding the dimension
img_tensor /=255.0
print(img_tensor.shape)
```

```python
# visualizing picture
import matplotlib.pyplot as plt
plt.imshow(img_tensor[0])       # only one image present in tensor  ,index -> 0
plt.show()
```

creating a Keras model that takes batches of images as input, and outputs the activations of all convolution and pooling layers.
use the Keras class Model . A model is instantiated using two arguments: an input tensor (or list of input tensors) and an output tensor (or list of output tensors). its gives multiple output

```python
from keras import models
layer_outputs = [layer.output for layer in model.layers[:8]] #extracting the feature of top 8 layer
activation_model = models.Model(inputs =model.input ,outputs = layer_outputs) # creates a model that will return these output ,given the input
```

When fed an image input, this model returns the values of the layer activations in the
original model. return multiple output
one input -> 8 output (one per layer)

```python
activations = activation_model.predict(img_tensor)  # return activation of 1st eight layer 
```

```python
first_layer_activation =activations[0] 
print(first_layer_activation.shape)
```

```python
print('shape of first eight layer activation return by activation model :\n')
for i in range(0,8):
    print(activations[i].shape)
```

```python
for i in range(0 ,5):
    plt.matshow(first_layer_activation[0,: ,: ,i] ,cmap ='viridis')
```

As we can see the different channel of second layer activation encode generic part of images like edges ,diagonal edges ....

```python
#Visualizing the 20 channel of 2nd activation
second_layer_activation =activations[1]
for i in range(0 ,5):
    plt.matshow(second_layer_activation[0,: ,: ,i] ,cmap ='viridis')
```

As we can see the different channel of second layer activation encode more abstract part of images like cats eyes ,ears,....                                                                                                          

The sparsity of the activations increases with the depth of the layer: in the first
layer, all filters are activated by the input image; but in the following layers,
more and more filters are blank. This means the pattern encoded by the filter
isn’t found in the input image
